```python
# parse salary and calculate average salary from low and high end
df6 = df5.apply(parse_salary, axis=1)
```

```python
df7 = standard_skills(df6)
```

```python
df7_exploded = df7.explode('skills_clean')
df7_exploded = df7_exploded.dropna(subset=['skills_clean'])

# Apply the function on the 'skills' column to group similar skills
grouped_skills = skills_fuzzymatch(df7_exploded['skills_clean'], threshold=80)

# Add a new column for grouped skills
df7_exploded['grouped_skill'] = df7_exploded['skills_clean'].apply(lambda row: assign_group(row, grouped_skills))
```

```python
df7.to_csv(f'{path}12172024_linkedin_data_visualization.csv', index=False)
df7_exploded.to_csv(f'{path}12172024_linkedin_data_exploded.csv', index=False)
```
