```python
import pandas as pd
import uuid
from collections import Counter
```

```python
df = pd.read_csv('Job intention survey.csv')
df.head(3)
```

```python
df['survey_id'] = [str(uuid.uuid4()) for _ in range(len(df))]

df_id = df[['Email Address','First Name','Last Name','Student ID','survey_id']]
df = df.drop(columns=['Timestamp','Email Address','First Name','Last Name','Student ID'])

df.head(3)
```

```python
job_titles = df['Job Title 1'].tolist() + df['Job Title 2'].tolist() + df['Job Title 3'].tolist()
job_titles2 = [title for entry in job_titles if isinstance(entry, str) for title in entry.split('/')]
job_titles3 = [title.split('-')[0].strip().capitalize() for title in job_titles2]
job_titles3
```

```python
title_counter = Counter(job_titles3)
print(title_counter)
```
